## Document loaders
**There are tons of document loaders** that converts any pdf, apis, json etc. document to langchain document object so to work with langchain

### Text Loader

```python
from langchain_community.document_loaders import TextLoader

path = "requirements.txt"
loaders = TextLoader(path)
docs = loaders.load()
print(docs)
print(len(docs))
print(type(docs))
print(docs[0].page_content)
```

### PyPDFLoader
**It's aweosome as it loads a full pdf pg by pg where each pg act as a separate page**

```python
from langchain_community.document_loaders import PyPDFLoader
path = "paper.pdf"
pdfload = PyPDFLoader(path)
docs = pdfload.load()
print(len(docs))
print(type(docs))
print(docs[4].page_content)
```

#### For multiple pdf files

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader

multi_load = PyPDFDirectoryLoader(path="pdf",glob="**/*.pdf") #Glob for multiple documents along
docspd = multi_load.load()

print(type(docspd))
print(len(docspd))
print(docspd[13].page_content)  # It works because they have 6 + 10 total 16 pages -_-
```

## CSVLoader

```python
from langchain_community.document_loaders import CSVLoader

path = ""
loader = CSVLoader(path)
docu = loader.load()
print(docu)
```

## JSONLoader

```python
from langchain_community.document_loaders import JSONLoader

path = ""
document = JSONLoader(path,jq_schema=".",text_content=False)
doc = document.load()
print(doc)
```
