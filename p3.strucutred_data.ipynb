## Concept of Typing
### Adding types I want as a output
**You'll get everything in python crash course**


```python
from typing import List, Optional, Annotated, TypedDict

#Adding types I want as a output
class Movie(TypedDict):
    title:Annotated[str,"The title of the movie"]
    release_year:Annotated[int,"Year the movie was released"]
    genres:Annotated[List[str],"List of genres the movie belongs to"]
    rating:Annotated[float,"Average rating of the movie on scale 1 to 10"]
    box_office:Annotated[Optional[float],"Worldwide box office in millions USD if known"]
```

**Structured and Unstructured output**

```python
prompt = "Details of the movie Inception"
structured_llm = llm.with_structured_output(Movie)
response = structured_llm.invoke(prompt)
print(response)
print(type(response))
print("\n\n\n\n")

#Getting unstructure output
ans = llm.invoke(prompt)
print(ans.content)
print(ans.type)
```

## Learning Pydantic
**Used to validate data, Industry grade, description, default values**
**The problem was Movie takes int/float values too which is not applicable for movie title.**

```python
from pydantic import BaseModel, Field
class Movie_Py(BaseModel):
    title:str=Field(...,description="Name of the movie")
    release_year:int=Field(..., description="Year the movie was released")
    genres:list[str] = Field(..., description="List of genres the movie belongs to")
    rating:float=Field(...,description="Average rating of the movie on scale 1 to 10")
    box_office:Optional[float]=Field(...,description="Worldwide box office in millions USD if known")

structured_pydantic_llm = llm.with_structured_output(Movie_Py)
response_py = structured_pydantic_llm.invoke(prompt)
print(response_py) #To get normal value
print(response_py.model_dump()) #To get dictionary type value
print(response_py.model_dump_json()) #To get json value
print(type(response_py))
```
