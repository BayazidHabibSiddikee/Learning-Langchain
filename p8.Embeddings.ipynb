## Starting with normal model

```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model="marin:latest")

text = "Hello wolrd what's up I'm Bayazid the future Millionare working on Langchain and I noticed something\n\
    that I started earlier I read hackings, networkings, IoT, banking, commerce, marketting strategy, psychology, market research, blockchain etc. and then\n\
        I came back to learn complete python and started with the automate the boring stuff using python and then 3/4 more books in python\n\
            learned about IoT AVR programming arduino esp programming, maths statistics in ML\n\
                and so I planned to go back and learn ML in C++ to make it fast and durable and now I have the ability who can combine multiple programming language to work on same table\n\
                    Thanks the frustrated loosing Bayazid who is working hard I'm proud of myself now that I can take any subject and has the ability to understand those because the back door is nothing ness"

response = llm.invoke(text)
print(response.content)
```

## Start embeddings

```python
from langchain_ollama import OllamaEmbeddings

embeder = OllamaEmbeddings(model="nomic-embed-text:latest")

vector = embeder.embed_query(text)
print(len(vector))
print(type(vector))
print(vector)
```

## Vectorise pdf file or list of texts

### Load PDF

```python
from langchain_community.document_loaders import PyPDFLoader

path = "paper.pdf"
pdf = PyPDFLoader(path).load()
print(pdf[1].page_content[:200])
```

### Split the PDF

**It's already a list**
recursive_split = RecursiveCharacterTextSplitter(
    chunk_size=100, chunk_overlap=9
).split_text(pdf[1].page_content)

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

recursive_split = RecursiveCharacterTextSplitter(
    chunk_size=100, chunk_overlap=9
).split_text(pdf[1].page_content)

print(len(recursive_split))
print(type(recursive_split))
print(recursive_split[:2000])
```

### Vectorise the list

```python
vector_list = embeder.embed_documents(recursive_split)
print(len(vector_list))
print(type(vector_list))
print(vector_list)
```
