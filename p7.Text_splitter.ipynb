# Text Splitter
**It's splitter that breaks text file in to chunks, and transfer to numerics and show in higher dimension**

### Loading PDF File to experiment with

```python
from langchain_community.document_loaders import PyPDFLoader

path = "paper.pdf"
pdf = PyPDFLoader(path).load()
print(pdf[1].page_content[:200])
```

## CharacterTextSplitter
**Split by character chunks, means no tracks, simple & fast**

```python
from langchain_text_splitters import CharacterTextSplitter

ch_splitter = CharacterTextSplitter(
    separator=" ", chunk_size=30, chunk_overlap=3
).split_text(pdf[1].page_content)

#print("\n\n\n\n")
print(len(ch_splitter))
print(type(ch_splitter))
print(ch_splitter)
for x in ch_splitter:
    print(x)
```

## RecursiveCharacterTextSplitter
**It's commonly used because it separates at natural boundaries like paragraph sentence word**
separator=["\n\n","\n","."," ",""] are used commonly

```python
from langchain_community.document_loaders import PyPDFLoader

path = "paper.pdf"
pdf = PyPDFLoader(path).load()
print(pdf[1].page_content[:200])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

recursive_split = RecursiveCharacterTextSplitter(
    chunk_size=100, chunk_overlap=9
).split_text(pdf[1].page_content)

print(len(recursive_split))
print(type(recursive_split))
print(recursive_split[:2000])

for x in recursive_split:
    print(x)
```
